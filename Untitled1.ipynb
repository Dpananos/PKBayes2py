```python
%load_ext autoreload
%autoreload 2
import numpy as np
import pandas as pd
import os
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
results = os.listdir('data/opts')

def read_file(x):
    
    with open('data/opts/'+x, 'rb') as file:
        data = pickle.load(file)
    return data


data = list(map(read_file, results))
names = map(lambda x: x[:-4], results)


df = pd.DataFrame(np.vstack(data).T, columns=names)
df.index.name = 'subject'
df = df.reset_index()


f = pd.melt(df, id_vars=['subject','theoretically_best_scores'], value_name = 'val').round(3)

f['delta'] = f.theoretically_best_scores - f.val
```

```python
f = f.query("variable != 'pop_best_scores'")

paper_names = ['No Covariate Model','No Covariate Model + 1 Sample', 'Covarite Model', 'Covarite Model + 1 Sample','Optimal Sampling Time', 'Q Learning']
actual_names = ['tdm_only_scores', 'one_sample_tdm_scores', 'covar_only_scores', 'one_sample_scores','opt_time_scores', 'q_learning_scores']

name_change = pd.DataFrame({'variable' : actual_names, 'Method': paper_names })

f = f.merge(name_change)
```

```python
fig, ax = plt.subplots(figsize=(8,8), dpi = 120)
order = ['tdm_only_scores','one_sample_scores','covar_only_scores','one_sample_tdm_scores','q_learning_scores','opt_time_scores']
sns.boxplot(data=f, 
            y='Method', 
            x='delta',
            ax=ax,
           order = paper_names,
           orient='h').set(xlabel = 'Differences Between Achieved and Best Loss')
```

```python
f.to_csv('data/generated_data/simulation_results.csv', index = False)
```

```python
from scipy.stats import sem
f.groupby(['Method']).agg(
    expected_score = pd.NamedAgg(column = 'val', aggfunc = np.mean),
    median_score = pd.NamedAgg(column = 'val', aggfunc = np.median),
    score_deviation =  pd.NamedAgg(column = 'val', aggfunc = np.std),
    
    
    expected_difference = pd.NamedAgg(column = 'delta', aggfunc = np.mean),
    median_differece = pd.NamedAgg(column = 'delta', aggfunc = np.median),
    difference_sd = pd.NamedAgg(column='delta',aggfunc = np.std)
).round(2).loc[['No Covariate Model','No Covariate Model + 1 Sample', 'Covarite Model', 'Covarite Model + 1 Sample','Optimal Sampling Time', 'Q Learning'],:]
```
